## <span style = 'color:blue'> Define Inputs and file paths as well as downloading fillings from EDGAR

```python
import os

# Define the company in put folder
company_directory_dict = {}
company_names = input("Tickers of Companies to be downloaded (e.g. AAPL, MSFT)")
company_names = [name.strip().upper() for name in company_names.split(',')]

for name in company_names:
    company_statement = os.path.join(name, 'financial_statement')
    os.makedirs(company_statement, exist_ok = True)
    company_directory_dict[name] = company_statement

from edgar import get_filings, Company, set_identity

# Use Edgar to download company filings
set_identity('pandavunguyen@gmail.com')

# Loop through companies list from input
company_filings_dict = {}
for name in company_names:
    company = Company(name)
    company_filings_dict[name] = {}
    filings = company.get_filings(form = '10-K')       
    company_filings_dict[name] = filings
```

## <span style = 'color:blue'> Reading the HTML Files 

```python
start_year = int(input("Enter start year (e.g., 2018): "))
end_year = int(input("Enter end year (e.g., 2023): "))

def download_10k_html(filings, company_folder, start_year, end_year):
    try:
        filing_year = int(filings.filing_date.strftime('%Y'))
        if start_year <= filing_year <= end_year:
            html_content = filings.html()
            if html_content:
                file_path = os.path.join(company_folder, f'{filing_year}.html')
                with open(file_path, 'w', encoding='utf-8') as f:
                    f.write(html_content)
                print(f'HTML downloaded at: {file_path}')
            else:
                print(f'Failed to read HTML contents for {filing_year}')
    except Exception as e:
        print(f'Error Downloading HTML files: {e}')
    
for name in company_names:
    company_folder = company_directory_dict[name]
    for filings in company_filings_dict[name]:
        download_10k_html(filings, company_folder, start_year, end_year)
    print('')
```
